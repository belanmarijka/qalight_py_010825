# Домашня робота: інструкції до виконання

1. Створіть нову гілку, яка буде використовуватись для змін, за допомогою команди `git checkout -b homework04`.
2. Відкрийте файл `homework_04.py` та виконайте ДЗ згідно з інструкціями написаними в коментарях до коду.
3. Перемістіть файл у репозиторій, створіть папку `lesson_04`  та помістіть файл у папці.
4. Додайте внесені зміни до стейджингу за допомогою команди `git add lesson_04/homework_04.py`
8. Зробіть коміт з змінами, додавши опис виконаних змін за допомогою команди `git commit -m "Виконані зміни"`.
9. Відправте свої зміни до вашого форку репозиторію за допомогою команди `git push`.
11. Створіть pull request у  власну гілку `main` на головний репозиторію [https://github.com](https://github.com/), натиснувши кнопку "New pull request" на відповідному розділі репозиторію.
    
    ![Untitled](https://i.ibb.co/Jjf4sHmd/git-pr-01.png)
    
12. Додайте викладача до рев’юверів
    
    ![Untitled](https://i.ibb.co/3YFH46bZ/git-rev.png)
    
13. **Посилання на PR вставите у форму відповіді для ДЗ в навчальній системі:**
    
    ![class](https://i.ibb.co/7dsVb8QC/firefox-n-ZDf-WHg-N5-R.png)
    
14. Після перевірки вашого pull request викладач може підтвердити, що все добре:
    
    ![Untitled](https://i.ibb.co/TxLZxZRb/git-rev-01.png)
    
15. Або вказати помилки та запропонувати змінити щось
    
    ![Untitled](https://i.ibb.co/PsTQVrcD/git-rev-02.png)
    
16. Після отримання Approved гілку можна замержити у `main`
